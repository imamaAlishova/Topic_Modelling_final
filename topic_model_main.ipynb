## Data Preprocessing 


### Data loading 

```python

# importing required modules 
import PyPDF2 
import pandas as pd
import os


pdfs_text=[]
path=r'C:\Users\imama\OneDrive\Desktop\TOP\special needs education'
for filename in os.listdir(path):
    files = os.path.join(path, filename)
# creating a pdf file object 
    pdfFileObj = open(files, 'rb') 
# creating a pdf reader object 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
# number of pages in pdf file
# creating a page object
    count=pdfReader.numPages
#     print(count)
    output=""
    for i in range(count):
        page_obj=pdfReader.getPage(i)
# extracting text from page 
        output+=page_obj.extractText()
#         print(output)
    pdfs_text.append(output)
    
len(pdfs_text)
# pdfs_text[3]
    
  
```

### Prepare Data for LDA Analysis
#### We need to prepare the input data for training LDA model
#### We start by tokenizing the text and removing stopwords




      

```python
# importing libraries 
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords



stop_words = stopwords.words('english') #stop words in english language



stop_words.extend(['doi','sen','et','al','education','teachers','teacher','support','parent','students','student','via','sd',
                    'ed','ie']) #list of the stop words

# this function removes punctuations 
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))

# this function removes stopwords        
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]


data = papers.paper_text_processed.values.tolist()#converting df to a list 
data_words = list(sent_to_words(data)) #listing the data cleaned from punctuation

# remove stop words
data_words = remove_stopwords(data_words) # final data cleaned from stop words

# print(len(data_words[1][1]))
# printing final data 
print(data_words[13][:9][:5])
print(stop_words)
```

#### We convert the tokenized object into a corpus and dictionary
   

```python
import gensim.corpora as corpora
# Create Dictionary
id2word = corpora.Dictionary(data_words)
id2word.filter_extremes(no_below=10, no_above=0.5, keep_n=50000)

# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:9][8][:30])
```

## LDA Training

```python
from pprint import pprint
# number of topics
num_topics = 20

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics,
                                       alpha='asymmetric',
                                       passes=40)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import pyLDAvis.gensim_models
import pickle 
import pyLDAvis
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('./results/lda'+str(num_topics))
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, './results/lda'+ str(num_topics) +'.html')
LDAvis_prepared
```

```python
topics_data = pd.DataFrame(columns=['Topic','Words'])

# Get the topics and their associated words
for topic_id, topic in lda_model.print_topics(num_topics=-1, num_words=10):
    topic_words = topic.split('+')
    topic_words = [word.split('*')[1].split('"')[1] for word in topic_words]
    topics_data.loc[topic_id] = [topic_id, topic_words]
```

```python
num_words = 10
num_topics = 20
import matplotlib.pyplot as plt
topics_data2 = topics_data.set_index('Topic').Words.apply(pd.Series).stack()
topics_data2 = topics_data2.reset_index()
topics_data2.columns = ['Topic','word_index','Words']
topics_data2.drop('word_index',axis=1,inplace=True)

df_grouped = topics_data2.groupby(['Topic', 'Words']).size().reset_index(name='frequency')

# calculate the total number of occurrences of each word throughout all the topics
df_grouped = df_grouped.groupby(['Words']).sum().reset_index()
df_grouped.drop('Topic',axis=1,inplace=True)

merged_df = pd.merge(topics_data2, df_grouped, on='Words')
merged_df['frequency'] = merged_df['frequency']-1
grouped_df = merged_df.groupby(['Topic']).sum().apply(lambda x: (x/((num_topics-1)*num_words)))
grouped_df.plot(kind='bar', y='frequency',ylabel='Words Similarity%', rot=0,figsize=(10,5))
plt.show()
```
